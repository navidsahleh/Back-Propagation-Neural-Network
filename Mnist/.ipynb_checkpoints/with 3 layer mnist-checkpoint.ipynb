```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
%matplotlib inline
df = pd.read_csv('test.csv')
df.describe()
```

```python
label= pd.get_dummies(df['label'])
df = pd.concat([df,label],axis=1)
df.drop(['label'],inplace=True,axis=1)
#df_norm.head()
df.describe()
```

```python
data=df.to_numpy()
```

```python
#activation function (sigmoid)
def sigmoid(x):
    return 1.0/(1.0+np.exp(-x))
#derivation of activation function(sigmoid)
def der_sigmoid(x):
    return x * (1.0-x)
#activation function
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def relu(X):
    return np.maximum(0,X)
def der_relu(X):
    return np.where(X <= 0,0,1)
#softmax function
def softmax(X):
    expo = np.exp(X)
    expo_sum = np.sum(np.exp(X))
    return expo/expo_sum
```

```python
#input layer
X_train = data[:1000,:784]
X_train.shape
```

```python
#input layer
y_train = data[:1000,784:]
y_train.shape
```

```python
#weights for first and second layer
w1_1 = np.random.uniform(-1,1,(784,100))
#w1_1= 2 * np.random.random((784,7)) -1
```

```python
#weights for third and fourth layer
w2_2 = np.random.uniform(-1,1,(100,10))
#w3_3=2 * np.random.random((6,10)) -1
```

```python
w1=np.copy(w1_1)
w2=np.copy(w2_2)
```

```python
batch_size=20
lis = []
for i in range(30):

    #forward propagation
    for j in range(0,len(X_train),batch_size):
        
        k = X_train[j:j+batch_size]
        p = y_train[j:j+batch_size]
        
    
        #second layer output with 3 neuron
        z2=np.dot(k,w1)
        z2=(z2 - z2.min())/(z2.max() - z2.min())
        a2=tanh(z2)
    
        #third layer output with 2 neuron

        z3=np.dot(a2,w2)
        a3=tanh(z3)

    
    
        #Backpropagation
    
    #    cost = (abs(y_train - a4)).mean()
        cost = 0.5 * (p - a3)**2
        #lis.append(np.sum(cost))
    
        #print('train',i,':',np.sum(cost))
        #derviation of cost function for w3
        delta3= (p - a3) * (1-tanh(a3)**2)
        gamma2=np.dot(a2.T,delta3)
    
    
        delta2 = np.dot(delta3,w2.T) * (1-tanh(a2)**2)
        gamma1 = np.dot(k.T,delta2)

       
    
        #updating the weights

        #learning rate
        q=0.1

    
        w2 += (q*gamma2)
        w1 += (q*gamma1)
        


print('the prediction:\n',a3[:4])
print('\n the output:\n',y_train[:4])
print('the cost:\n',np.sum(cost))
```
